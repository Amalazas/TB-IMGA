```python
import numpy as np
```

```python
vars = np.array([10 * np.random.rand(10) for _ in range(10)] + [-10 * np.random.rand(10) for _ in range(3)])
```

```python
def rank_outliers(vars):
        variables_mean = vars.mean(axis=0)
        ranked_sol = sorted(vars, key=lambda var: np.dot(var, variables_mean))
        
        print("\n\nRANKING")
        for i, sol in enumerate(ranked_sol):
            print(f"{i}. ({np.dot(sol, variables_mean)}): {sol}")        
        
        return ranked_sol
```

```python
rank_outliers(vars)
```
